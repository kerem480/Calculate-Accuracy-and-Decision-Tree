```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
from sklearn import tree
```

```python
df=pd.read_csv("C:\\Users\\90533\\Desktop\\winequalityN.csv",encoding =('ISO-8859-1'),low_memory =False)
```

```python
df.head()
```

```python
df[["volatile acidity","fixed acidity"]]
```

```python
df = df.drop(columns=['citric acid', 'residual sugar',"chlorides","free sulfur dioxide","total sulfur dioxide",
                     "density","pH","sulphates","alcohol","quality"])
```

```python
data=df[["fixed acidity","volatile acidity","type"]][:100]
```

```python
df=df.dropna()
data=data.dropna()
```

```python
x=data.drop(columns=["type"])
y=data["type"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8)

model=DecisionTreeClassifier()
model.fit(x_train,y_train)
predictions=model.predict(x_test)

score=accuracy_score(y_test,predictions)
score
```

```python
x=df.drop(columns=["type"])
y=df["type"]

model=DecisionTreeClassifier()
model.fit(x,y)

tree.export_graphviz(model,out_file="wine-quality.dot",feature_names=["fixed acidity","volatile acidity"],class_names=sorted(y.unique()),
                    label="all",
                    rounded=True,
                    filled=True)
```
